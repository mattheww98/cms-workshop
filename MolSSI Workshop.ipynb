```python
important_lines = []
energies = {}
energies['Total Energy'] = 0
with open(ethene_file, "r") as saptout:
    for line in saptout:
        if 'Electrostatics  ' in line:
            important_lines.append(line)
        if 'Exchange  ' in line:
            important_lines.append(line)
        if 'Induction  ' in line:
            important_lines.append(line)
        if 'Dispersion  ' in line:
            important_lines.append(line)
for line in important_lines:
    words = line.split()
    comp = words[0]
    energy = words[3]+' kcal/mol'
    energies[comp] = energy
    energies['Total Energy'] += float(words[3])
energies['Total Energy'] = str(energies['Total Energy'])+' kcal/mol'
for key in energies:
    print(key+" : "+energies[key])
```

```python
datafile = open('energies.txt', 'w+') #open file for writing, create if needed
for f in filenames:
    # Get molecule name
    file_name = os.path.basename(f)
    split_name = file_name.split(".")
    molecule_name = split_name[0]
    
    #Read in data from file
    outfile = open(f,'r')
    data = outfile.readlines()
    outfile.close()
    
    #Loop through data for energies
    for line in data:
        if 'Final Energy' in line:
            energy_line = line
            words = energy_line.split()
            energy = float(words[3])
            datafile.write(F'{molecule_name} \t {energy} \n')
            #F to convert variable to string, t for tab to align, n for newline
datafile.close()
```

```python
water_file = os.path.join('data', 'water.xyz')
dataset = numpy.genfromtxt(fname = water_file, skip_header=2, dtype = 'unicode')
no_rows = len(dataset[:,0])
bond_lengths = {}
for i in range(no_rows):
    row = dataset[i,:]
    atom = row[0]
    coords = row[1:].astype(float)
    for j in range(i,no_rows):
        row2 = dataset[j,:]
        atom2 = row2[0]
        coords2 = row2[1:].astype(float)
        atoms = atom + ' to ' + atom2
        distances = ()
        square_sum = 0
        for k in range(1,3):
            distance = coords2[k]-coords[k]
            square_sum+= distance**2
        bond = numpy.sqrt(square_sum)
        if bond > 0 and bond <= 1.5:
            bond_lengths[atoms] = bond
with open("bond_lengths.txt", "w+") as output_file:
    for key in bond_lengths:
        output_file.write(F'{key} : {bond_lengths[key]} \n')
```
